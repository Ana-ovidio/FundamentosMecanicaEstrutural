```python
import matplotlib.pyplot as plt
from anastruct import SystemElements
```

```python
ss = SystemElements()
```

### **1.4**


```python
nodes = {
    'A': (0,0),
    'B': (2,0), 
    'C': (4,0),
    'D': (6,0),
    'E': (4, 1.5),
    'F': (2,3),
    'G': (0,3)
}

nodes_names = list(nodes.keys())
cncts = [('A', 'B'), ('A', 'F'), ('A', 'G'),
         ('B', 'C'), ('B', 'F'), ('B', 'E'), 
         ('C', 'E'), ('C', 'D'), ('D', 'E'), 
         ('E', 'F'), ('F', 'G')]
```

```python
for connection in cncts:
    element_truss = [list(nodes.get(connection[0])), list(nodes.get(connection[1]))]
    ss.add_truss_element(location = element_truss)
```

```python
reation_A = ss.find_node_id(list(nodes.get('A')))
reation_G = ss.find_node_id(list(nodes.get('G')))

ss.add_support_roll(node_id=reation_A, direction=1, angle=-90)
ss.add_support_hinged(node_id=reation_G)

fig = ss.show_structure(scale = 0.75, show=False)
plt.grid(linestyle='--')
plt.show()
```

```python
extern_force =- 5000
```

```python
for node in ['B', 'C', 'D']:
    extern_force_loc = ss.find_node_id(list(nodes.get(f'{node}')))
    ss.point_load(node_id=extern_force_loc, Fy=extern_force)
```

```python
ss.solve()
```

```python
fig = ss.show_structure(scale = 0.75, figsize=(8,8), show=False)
plt.grid(linestyle='--')
plt.show()
```

```python
fig = ss.show_reaction_force(scale = 0.75, figsize=(8,8), show=False)
plt.grid(linestyle='--')
plt.show()
```
